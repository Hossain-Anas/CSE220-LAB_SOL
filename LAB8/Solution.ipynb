```python
class BTNode:
  def __init__(self, elem):
    self.elem = elem
    self.right = None
    self.left = None
```

```python
def inorder(root):
  if root == None:
    return

  inorder(root.left)
  print(root.elem, end = ' ')
  inorder(root.right)
```

```python
def tree_construction(arr, i = 1):
  if i>=len(arr) or arr[i] == None:
    return None
  p = BTNode(arr[i])
  p.left = tree_construction(arr, 2*i)
  p.right = tree_construction(arr, 2*i+1)
  return p


root2 = tree_construction([None, 'A', 'B', 'E', 'C', 'D', 'F', 'G', None])
inorder(root2)
```

```python
def convert_mirror(root):
    if root == None:
        return None
    
    new_root = BTNode(root.elem)
    temp = root.right
    new_root.right = convert_mirror(root.left)
    new_root.left = convert_mirror(temp)

    return new_root


# DRIVER CODE
root = BTNode(10)
n1 = BTNode(20)
n2 = BTNode(30)
n3 = BTNode(40)
n4 = BTNode(60)

root.left = n1
root.right = n2

n1.left = n3
n1.right = n4

print("Given Tree Inorder Traversal: ", end=" ")
inorder(root)  # Given Tree Inorder Traversal:  40 20 60 10 30
print()

root2 = convert_mirror(root)
print("Mirrored Tree Inorder Traversal: ", end=" ")
inorder(root2)  # Mirrored Tree Inorder Traversal:  30 10 60 20 40
```

```python
def smallest_level(root):
    level = {}

    def check_min(root, key=0):
        if root is None:
            return

        if key not in level:
            level.update({key: root.elem})

        elif key in level:
            if level[key] > root.elem:
                level.update({key: root.elem})

        check_min(root.left, key + 1)
        check_min(root.right, key + 1)
    

    check_min(root)
    return level


# DRIVER CODE

root = tree_construction([None, 4, 9, 2, 3, -5, None, 7])
print("Given Tree Inorder Traversal: ", end=" ")
inorder(root) # Given Tree Inorder Traversal:  3 9 5 4 2 7

print()

print("Level Wise Smallest Value: ", end=" ")
print(smallest_level(root))  # Level Wise Smallest Value:  {0: 4, 1: 2, 2: -5}
```

```python
def sum_tree(root):
    res = [True]

    def my_sum(root):
        
        if root is None:
            return 0
        
        if root.left is None and root.right is None:
            return root.elem
        
        l = my_sum(root.left)
        r = my_sum(root.right)

        if l+r != root.elem :
            res[0] = False
            return -1
        
        return l+r+root.elem

    my_sum(root)
    
    return res[0]

root1 = tree_construction([None, 26, 10, 3, 4, 6, None, 3])
res = sum_tree(root1)
print(res)
```
