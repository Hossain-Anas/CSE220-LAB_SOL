```python
#You must run this cell to print matrix and for the driver code to work
def print_matrix(m):
  row,col = m.shape
  for i in range(row):
    c = 1
    print('|', end='')
    for j in range(col):
      c += 1
      if(len(str(m[i][j])) == 1):
        print(' ',m[i][j], end = '  |')
        c += 6
      else:
        print(' ',m[i][j], end = ' |')
        c += 6
    print()
    print('-'*(c-col))
```

#Zigzag Walk

```python
def walk_zigzag(floor):
    row, column = floor.shape
    col = "even"

    for i in range(column):
        if col == "even":
            for j in range(row):
                if j%2 == 0 :
                    print(floor[j][i], end=" ")
            print()
            col = "odd"

        elif col == "odd" :
            for j in range(column-1,-1,-1):
                if j%2 != 0 :
                    print(floor[j][i],end =" ")

            print()
            col = "even"


floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0'],
                  ['1' , '4' , '2' , '8' , '6']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9 1
# 1 2
# 4 7 2
# 4 9
# 1 8 6
print('################')
floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9
# 1 2
# 4 7
# 4 9
# 1 8
```

#Wall Up Trost District

```python
def create_fence(district, depth):
    district_row, district_col = district.shape

    prev_row, prev_col = district_row, district_col
    new_row = prev_row+(depth*2)
    new_col = prev_col + (depth*2)
    new_matrix = np.full((new_row,new_col),8)
    for i in range(prev_row):
        for j in range(prev_col):
            new_matrix[i+depth][j+depth] = district[i][j]

    return new_matrix

depth = 1
district = np.array([[2,3,4], [3,4,6], [2,1,4]])
print_matrix(district)
ans = create_fence(district, depth)
print_matrix(ans)
#This will print
# |  8  |  8  |  8  |  8  |  8  |
# -------------------------------
# |  8  |  2  |  3  |  4  |  8  |
# -------------------------------
# |  8  |  3  |  4  |  6  |  8  |
# -------------------------------
# |  8  |  2  |  1  |  4  |  8  |
# -------------------------------
# |  8  |  8  |  8  |  8  |  8  |
# -------------------------------
print('################')
depth = 2
district = np.array([
                 [2,3,4,1],
                 [3,4,6,5],
                 [2,1,4,7]
                ])
print_matrix(district)
ans = create_fence(district, depth)
print_matrix(ans)
```

#Crows vs Cats

```python
def strength_difference(clubroom):
    row,col = clubroom.shape
    array_size = int(((row-1)*row)/2)
    strength_diff = np.array([0]*array_size)
    index = 0

    for i in range(row-1):
        for j in range(col):
            val = clubroom[i][j]

            if i==j or val==-1 :
                continue
            
            else:
                conjugate = clubroom[j][i]
                res = val-conjugate
                strength_diff[index] = res
                index += 1
                clubroom[j][i] = -1
    
    return strength_diff

clubroom = np.array([
[1,  2,  9,  7],
[4,  5,  1,  8],
[3,  6,  2,  7],
[2,  8,  6,  3]
])
print_matrix(clubroom)
returned_value = strength_difference(clubroom)
print('Strength Difference Array is : ', returned_value)
unittest.output_test(returned_value, np.array([-2, 6, 5, -5, 0, 1]))
```

#ATM's Triangle

```python
def atm_triangle(n):
    new_array = np.zeros((n,n),dtype=int)
    for i in range(n):
        for j in range(i+1):
            if j==0 or j==i :
                new_array[i][j] = i+1

            else :
                new_array[i][j] = -1
    
    if n<=2 :
        return new_array
    
    for i in range(2,n):
        sum = 0
        for j in range(i+1):
            sum += new_array[i-1][j]
            if new_array[i][j] == -1 :
                new_array[i][j] = sum
    
    return new_array
    


def print_without_none(matrix):
    row,col = matrix.shape

    for i in range(row):
        for j in range(col):
            if matrix[i][j] != 0 :
                print(matrix[i][j], end = "  ")
        
        print()


n = 5
returned_value = atm_triangle(n)
print_without_none(returned_value)
#This should print
# 1
# 2  2
# 3  4  3
# 4  7  10  4
# 5  11 21  25  5
```

#Trace The BOT

```python
def moving_around(cmds):
    
    grid = np.full((7, 7), '.')
    row, col = 3, 3
    grid[row][col] = '-'

    for i in cmds:
        new_row = row
        new_col = col
        if i == 1:
            row -= 2
            col -= 2
        elif i == 2:
            row -= 2
            col -= 1
        elif i == 3:
            row -= 3
            col -= 3
        elif i == 4:
            row -= 1
            col -= 2
        elif i == 5:
            row -= 2
            col += 1
        elif i == 6:
            row -= 2
            col += 3
        elif i == 7:
            row -= 3
            col += 2
        elif i == 8:
            row -= 1
            col += 2
        elif i == 9:
            row += 2
            col -= 3
        elif i == 10:
            row += 2
            col -= 1
        elif i == 11:
            row += 1
            col -= 2
        elif i == 12:
            row += 3
            col -= 2

        if 0 <= row <= 6 and 0 <= col <= 6:
            grid[row][col] = '*'
            new_row = row
            new_col = col
        else:
            grid[new_row][new_col] = '/'

    return grid



cmds = np.array([5,11,2,9])
result = moving_around(cmds)
print_matrix(result)
#This should print
# -------------------------------------------
# |  .  |  /  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  *  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  *  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  -  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
```
