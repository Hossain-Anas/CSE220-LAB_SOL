```python
# Test 01: Play Right
def rightRotate(a) :
    temp = a[len(a)-1]
    for i in range(len(a)-1,0,-1):
        a[i] = a[i-1]
    a[0] = temp

def playRight(sequence, beats):

    for i in range(len(beats)):
        if beats[i] == 1:
            rightRotate(sequence)
    
    return sequence

print("///  Test 01: Play Right  ///")
sequence = np.array([10, 20, 30, 40, 50, 60])
beats = np.array([1, 0, 0, 1, 0, 1])
returned_value = playRight(sequence, beats)
print(f'Task 1: {returned_value}')  # This should print [40, 50, 60, 10, 20, 30]
```

```python
# Test 02: Discard Cards
def leftRotate(a,start,end) :
    temp = a[start]
    for i in range(start,end-1):
        a[i] = a[i+1]
    
    a[end-1] = temp

def discardCards(cards,number):
    i = 0;
    end = len(cards) 

    while i<end:
        if cards[i] == 2:
            cards[i] = 0
            leftRotate(cards,i,end)
            end-=1
            continue
        else:
            i+=1
            
    return cards


print("///  Test 02: Discard Cards  ///")
cards = np.array([1,2,3,2,8,2,2,5,7])
returned_value = discardCards(cards, 2)
print(f'Task 2: {returned_value}') # This should print [1, 3, 8, 5, 7, 0, 0, 0, 0]
unittest.output_test(returned_value, np.array([1, 3, 8, 5, 7, 0, 0, 0, 0]))
```

```python
# Test 04: Balance your Salami
def balanceSalami(salami):
    flag = False
    i = 0
    sum_i = 0

    while flag!=True and i<len(salami) :
        sum_j = 0
        sum_i += salami[i]

        for j in range(i+1,len(salami)):
            sum_j += salami[j]
        
        if sum_i == sum_j :
            flag = True

        i+=1

    return flag

print("///  Test 04: Balance Your Salami  ///")
salami = np.array([1, 1, 1, 2, 1])
returned_value = balanceSalami(salami)
print(f'Task 4: {returned_value}')  # This should print True
unittest.output_test(returned_value, True)

salami = [2, 1, 1, 2, 1]
returned_value = np.array(balanceSalami(salami))
print(f'Task 4: {returned_value}')  # This should print False
unittest.output_test(returned_value, False)

salami = [10, 3, 1, 2, 10]
returned_value = np.array(balanceSalami(salami))
print(f'Task 4: {returned_value}')  # This should print True
unittest.output_test(returned_value, True)
```

```python
# Test 05: Protecc Salami
def sort(a):
    for i in range(0,len(a)-1):
        for j in range(i+1,len(a)):
            if a[i] > a[j]:
                temp = a[j]
                a[j] = a[i]
                a[i] = temp

def count(value,array):
    count = 0
    for i in range(len(array)):
        if array[i] == value:
            count += 1
    
    return count

def unique_array(a,x):
    if len(a) == 0:
        return
    
    x[0] = a[0]
    i = 1
    j = 1
    while i<len(a):
        if a[i] != x[j-1] :
            x[j] = a[i]
            j+=1

        i+=1

    last_valid_index = j
    return last_valid_index

def protectSalami(salami):

    sort(salami)
    unique_salami = np.array([0]*len(salami))
    size_of_ua = unique_array(salami,unique_salami)

    flag = False
    i = 0

    while i<size_of_ua and flag!=True:

        count_i = count(unique_salami[i],salami)

        if count_i>1:

            for j in range(i+1,size_of_ua):
                count_j = count(unique_salami[j],salami)
                

                if count_i == count_j :
                    flag = True
                    break
        
        i+=1
      
    return flag

print("///  Test 05: Protecc Salami  ///")
salami = np.array([4,5,6,6,4,3,6,4])
returned_value = protectSalami(salami)
print(f'Task 5: {returned_value}') # This should print True
unittest.output_test(returned_value, True)

salami = np.array([3,4,6,3,4,7,4,6,8,6,6])
returned_value = protectSalami(salami)
print(f'Task 5: {returned_value}') # This should print False
unittest.output_test(returned_value, False)
```

```python

def mean(a):
    sum = 0
    for i in range(len(a)):
        sum += a[i]
    
    mean = sum/len(a)
    return mean

def standard_deviation(a):
    mn = mean(a)
    sum = 0

    for i in range(len(a)):
        temp = (a[i]-mn)**2
        sum += temp
    
    dev = (sum/(len(a)-1))**(1/2)
    dev = round(dev,2)
    return dev

def idk(a) :
    mean_a = mean(a)
    dev = standard_deviation(a)

    low = mean_a - (1.5*dev)
    high = mean_a + (1.5*dev)

    count = 0

    for i in range(len(a)):
        val = a[i]
        if val<low or val>high :
            count += 1


    if count == 0 :
        return None
    else :
        arr = np.array([0]*count)
        j = 0

        for i in range(len(a)):
            val = a[i]
            if val<low or val>high :
                arr[j] = val
                j+=1
                
        return arr,mean_a,dev
        

a = np.array([10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4])
new_array,mean,dev = idk(a)
print("The mean of the numbers is:", mean)
print("The standard deviation is:",dev)
print("New array:",new_array)
```
